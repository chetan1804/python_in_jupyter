```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
r_sum = np.sum(s, axis=1)
r_sum
```

```python
c_sum = np.sum(s, axis=0)
c_sum
```

```python
s[0:3,0:3]
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        n=s[i:i+3,j:j+3]
        print(n.sum())
```

```python
# column : [Age, Math Marks, Science Marks]
data = np.array([
    [18,85,78],
    [19,92,88],
    [17,76,95],
    [18,65,70],
    [20,90,85]
])
```

```python
# Find shape of the matrix
data.shape
```

```python
# Find the average age of students
np.mean(data[:,0])
```

```python
# Extract Math marks of all students
data[:,1]
```

```python
# Find highest Science mark
data[:,2].max() #Method 1
np.max(data[:,2]) # Method 2
```

```python
# Get details of the studen who scored more than 90 in maths
bool_index = data[:,1] >= 90
data1 = data[bool_index]
data1
```

```python
# Increase math marks of all students by 5
data[:,1] += 5
data
```

```python
# Find how many students are younger than 19
data1 = data[:,0] < 19
len(data[data1])
```

```python
# Calculate the average markes in each subject
np.mean(data[:,1:],axis=0)
```

```python
# Get data of students who scored at least 80 in both subject
data[(data[:,1] >=80) & (data[:,2] >= 80)]
```

```python
#  Replace all Science marks < 75 with 0
data[:,2][data[:,2]< 75] = 0
data
```
