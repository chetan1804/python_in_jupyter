# Matplotlib Library

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y = [5,6,7,8]
c="r"
plt.plot(x,y,c)
plt.show()
```

```python
x=[5,7,2,9,5]
y = [1,4,3,10,3]
c="r"
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
plt.plot(y, marker = 'o')
plt.show()
```

# Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.barh(x,y,color = 'hotpink')
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.barh(x,y,color = 'hotpink',height=0.5)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.bar(x,y,color = 'green',width=0.5)
plt.show()
```

```python
counts = [577,314]
labels = ['Male','Female']
colors = ['skyblue','lightpink']
explode = (0.1,0) # Male should be display seperatly

plt.figure(figsize=(6,6))
plt.pie(counts,labels=labels, autopct='%1.1f%%',startangle=140,colors=colors, explode=explode, shadow=True)

plt.title('Gender Distribution in Titanic')
plt.show()
```
