# Matplotlib Library

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y = [5,6,7,8]
c="r"
plt.plot(x,y,c)
plt.show()
```

```python
x=[5,7,2,9,5]
y = [1,4,3,10,3]
c="r"
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
plt.plot(y, marker = 'o')
plt.show()
```

# Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.barh(x,y,color = 'hotpink')
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.barh(x,y,color = 'hotpink',height=0.5)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
plt.bar(x,y,color = 'green',width=0.5)
plt.show()
```

```python
counts = [577,314]
labels = ['Male','Female']
colors = ['skyblue','lightpink']
explode = (0.1,0) # Male should be display seperatly

plt.figure(figsize=(6,6))
plt.pie(counts,labels=labels, autopct='%1.1f%%',startangle=140,colors=colors, explode=explode, shadow=True)

plt.title('Gender Distribution in Titanic')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
plt.plot(x,y)
plt.title("First Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Subplot

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
```

# Object Oriented Way

```python
fig = plt.figure(figsize=(8,6))
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)

axis2 = fig.add_axes([0.2,0.2,0.8,0.8])
axis2.plot(x,y)
```

# Type of Plots

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

# Save Plot

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("saveplt.png")
```
