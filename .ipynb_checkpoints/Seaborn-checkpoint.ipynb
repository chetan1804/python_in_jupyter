```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [8,7,6,5,4,3,2,1]
sns.lineplot(x=var1,y=var2)
plt.show()
```

```python
df1 = sns.load_dataset("penguins")
df1
```

```python
sns.lineplot(x='bill_length_mm', y="flipper_length_mm", data = df1)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y="flipper_length_mm",data=df1,hue="sex")
plt.plot()
```

```python
sns.lineplot(x='bill_length_mm', y="flipper_length_mm",data=df1,hue="sex",palette="rocket_r", style="sex", markers=["o",">"])
plt.plot()
```

```python
sns.barplot(x="island" , y="bill_length_mm", data=df1 , hue='sex', palette='prism')
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230])
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230],color="red")
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"],kde=True,bins=[170,180,190,200,210,220,230])
plt.show()
```
