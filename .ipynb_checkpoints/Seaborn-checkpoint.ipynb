```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.histplot(df['total_bill'], bins = 15, kde = True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins = 15, kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 15, kde = True)
plt.show()
```

```python
sns.jointplot(x = "total_bill", y="tip", data = df, kind="scatter")
```

```python
sns.jointplot(x = "total_bill", y="tip", data = df, kind="kde")
```

```python
sns.jointplot(x = "total_bill", y="tip", data = df, kind="hex")
```

```python
sns.jointplot(x = "total_bill", y="tip", data = df, kind="reg")
```

```python
sns.pairplot(df, hue="sex")
```

```python
sns.rugplot(df)
```

```python
sns.boxplot(x='tip' , y= 'day', data=df, palette="rainbow")
```

```python
sns.violinplot(x = 'day', y='total_bill', data = df)
```

# Striplot and Swarmplot

```python
sns.stripplot(x="day", y='total_bill', data=df)
```

```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

```python
# Combine violinplot and swarmplot
sns.violinplot(x='tip',y='day', data=df)
sns.swarmplot(x='tip',y='day', data=df)
```

# Matrixplot

# 1) Heat map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscor = tips[['total_bill', 'tip', 'size']]
tipscor
```

```python
tipscor.corr()
```

```python
sns.heatmap(tipscor.corr(), annot=True)
```

# 2) Cluster map

```python
sns.clustermap(tipscor.corr(), annot=True)
```

# 3) Pivot Table Heat Map

```python
pvtflights = flights.pivot_table(values = "passengers", index="month", columns = "year")
```

```python
pvtflights
```

```python
sns.heatmap(pvtflights)
```
