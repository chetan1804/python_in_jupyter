# Missing Data

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "A" : [1, 2, np.nan, 4, 5],
    "B" : [np.nan, 2,3,4,5],
    "C" : [1,2,3,np.nan,np.nan],
    "D" : [1, np.nan,np.nan,np.nan,5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
#Removing Missing Data
df.dropna()
```

```python
data = {
    "A" : [1, 2, np.nan, 4, 5],
    "B" : [np.nan, 2,3,4,5],
    "C" : [1,2,3,np.nan,np.nan],
    "D" : [1, np.nan,np.nan,np.nan,5]
}
df = pd.DataFrame(data)
```

```python
df1 = df.copy()
df1.dropna(thresh=3) # Remove row if 3 NaN in one row
```

```python
df1
```

```python
df1.fillna(0)
```

```python
values = {"A": 100, "B":200, "C":300,"D":400}
df1.fillna(value=values)
```

```python
df1.fillna(df1.mean())
```

# Pivot Tables

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Date": pd.date_range("2023-01-01", periods=20),
    "Product": ['A', 'B', 'C', 'D'] * 5,
    "Region": [
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South'
    ],
    "Sales": np.random.randint(100, 1000, 20),
    "Units": np.random.randint(10, 100, 20),
    "Rep": [
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice'
    ]
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df, values = "Sales", index = "Region", columns = "Product", aggfunc = "median")
```

```python
pivot2 = pd.pivot_table(df, values = ["Sales","Units"], index="Region", columns = "Product")
```

```python
pivot2
```

```python
pd.crosstab(df["Region"], df["Product"])
```
