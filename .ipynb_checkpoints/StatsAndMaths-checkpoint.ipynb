# Calculating Mean

```python
# Manual Way
population = [150,160,170,165,155]
sample = [150,170,165]
pop_mean = sum(population) / len(population)
sample_mean = sum(sample) / len(sample)
print("Population Mean" , pop_mean);
print("Sample Mean" , sample_mean)
```

```python
# Using Numpy library
import numpy as np

data = [85,90,92,88,75,98]
arithmatic_mean = np.mean(data)
print(arithmatic_mean)
```

```python
# Using statistic
import statistics as stat
var = [10,20,30,40,50]
stat_mean = stat.mean(var)
print(stat_mean)
```

```python
# from dataset
import seaborn as sns
import numpy as np

iris = sns.load_dataset('iris')
# iris.head()
mean_sepal_length = np.mean(iris['sepal_length'])
print("mean_sepal_length",mean_sepal_length)
```

# Median

```python
# Using numpy library
median_sepal_length = np.median(iris['sepal_length'])
print("median_sepal_length",median_sepal_length)
```

```python
# Using statistic
import statistics as stat
var = [10,20,30,40,50]
stat_median = stat.median(var)
print(stat_median)
```

```python
# Using pandas library
import pandas as pd

df = pd.DataFrame({
    'values' : [20,22,23,45,26,33]
})
print(df)
pandas_median = df['values'].median()
print('Median using pandas' , pandas_median)
```

# Mode

```python
# using statistic library
import statistics as stat
data = [85,78,80,90,64,55,78]

stat_mod = stat.mode(data)
print("Statistic mode",stat_mod)
```

```python
# Using Scipy library
from scipy import stats
data = [85,90,67,80,90]
sp_mode = stats.mode(data)
print(sp_mode)
```

# Variance

```python
# Using Numpy
import numpy as np
data = [2,4,4,5,6,7,8,9]

sample_variance = np.var(data , ddof=1)
pop_variance = np.var(data, ddof=0)

print("Sample variance", sample_variance)
print("Population Variance" , pop_variance)
```

```python
#using pandas
import pandas as pd

data = [2,4,4,5,6,7,8,9]
series = pd.Series(data)

sample_variance = series.var()
print("Sample variance", sample_variance)

pop_variance = series.var(ddof=0)
print("Population Variance" , pop_variance)
```

```python
# Using scipy
from scipy import stats

data = [2,4,4,5,6,7,8,9]

sample_variance = stats.tvar(data)
print("Sample variance", sample_variance)

pop_variance = stats.tvar(data,ddof=0)
print("Population Variance" , pop_variance)
```

# Standard Deviation

```python
# Using Numpy
import numpy as np

data = [60,65,70,75,80]

sample_std = np.std(data , ddof=1)
pop_std = np.std(data,ddof=0)

print("Sample Standard Deviation: ", sample_std)
print("Population Standard Deviation: ", pop_std)
```

```python
# Using Pandas
import pandas as pd

data = [80,60,65,70,75]
series = pd.Series(data)

sample_std = series.std(ddof=1)
pop_std = series.std(ddof=0)

print("Sample Standard Deviation: ", sample_std)
print("Population Standard Deviation: ", pop_std)
```

```python
# Using Scipy
from scipy import stats
import math

data = [80,60,65,70,75]

sample_std = math.sqrt(stats.tvar(data ,ddof=1))
pop_std = math.sqrt(stats.tvar(data,ddof=0))

print("Sample Standard Deviation: ", sample_std)
print("Population Standard Deviation: ", pop_std)
```

# Data Representation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# create a figure and axes
fig,axs = plt.subplots(3,2,figsize=(14,12))
fig.suptitle('Histograms with KDE of various distributions' , fontsize=16)

# Normal Distribution (Gaussian Distribution)
data_normal = np.random.normal(loc=0, scale=1, size=1000 )  # Mean=0 std_dev=1
sns.histplot(data_normal,kde=True, ax=axs[0,0] , color='blue', stat='density', bins=30)
axs[0,0].set_title('Normal Distribution')
axs[0,0].grid(True)



#Skewed Distribution
data_right_skewed = np.random.exponential(scale=1, size=1000) #Right Skewed
data_left_skewed = -np.random.exponential(scale=1, size=1000) #Left Skewed

#Right Skewed Histogram
sns.histplot(data_right_skewed,kde=True , ax=axs[0,1], color="red", stat='density', bins=30, label="Right Skewed")

#Left skewed histogram
sns.histplot(data_left_skewed,kde=True, ax=axs[0,1], color="green", stat="density", bins=30, label="Left Skewed")
axs[0,1].set_title("Skewed Distribution")
axs[0,1].legend()
axs[0,1].grid(True)


# Uniform Distribution
data_uniform = np.random.uniform(low=0,high=1,size=1000)   #Uniform distribution between 0 and 1
sns.histplot(data_uniform, kde=True, ax=axs[1,0], color='purple', stat="density", bins = 30)
axs[1,0].set_title('Uniform Distribution')
axs[1,0].grid(True)


#Bimodal Distribution
data_bimodal = np.concatenate([np.random.normal(loc=-2, scale=1, size=500),
                              np.random.normal(loc=2,scale=1,size=500)])
sns.histplot(data_bimodal,kde=True, ax=axs[1,1], color="orange", stat="density", bins=30)
axs[1,1].set_title("Bimodal Distribution")
axs[1,1].grid(True)



#Multimodal Distribution

data_multimodal = np.concatenate([
    np.random.normal(loc=-6, scale=1.5, size=500),
    np.random.normal(loc=0, scale=1, size=500),
    np.random.normal(loc=6, scale=1.5, size=500),
])

sns.histplot(data_multimodal, kde=True, ax=axs[2,0], color="brown", stat="density", bins=30)
axs[2,0].set_title("Multimodal Distribution")
axs[2,0].grid(True)

# Hide Empty subplot
axs[2,1].axis('off')

# Adjust Layout
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()


```
